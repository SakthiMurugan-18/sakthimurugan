```
import matplotlib.pyplot as plt
```

```
plt.figure(figsize=(10,5))
```

```
plt.figure(figsize=(10,5))

counts = df['MerchantID'].value_counts()

plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title("Merchant ID Distribution")
plt.show()
```

```
import matplotlib.pyplot as plt

counts = df['MerchantID'].value_counts().head(10)   # top 10 (recommended)

plt.figure(figsize=(10,5))
plt.bar(counts.index.astype(str), counts.values, color='skyblue')

plt.xlabel("Merchant ID")
plt.ylabel("Number of Transactions")
plt.title("Top 10 Merchant IDs by Transaction Count")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(10,5))

counts = df['MerchantID'].value_counts().sort_index()

plt.plot(counts.index, counts.values, marker='o')
plt.xlabel("Merchant ID")
plt.ylabel("Count")
plt.title("Merchant ID Distribution (Line Chart)")
plt.grid(True)

plt.show()
```

```
import matplotlib.pyplot as plt


counts_all = df['MerchantID'].value_counts()
counts_top10 = counts_all.head(10)


fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].pie(
    counts_all,
    labels=counts_all.index.astype(str),
    autopct='%1.1f%%',
    textprops={'fontsize': 8}
)
axes[0].set_title("Merchant ID Distribution")


axes[1].bar(
    counts_top10.index.astype(str),
    counts_top10.values,
    color='skyblue'
)
axes[1].set_title("Top 10 Merchant IDs")
axes[1].set_xlabel("Merchant ID")
axes[1].set_ylabel("Number of Transactions")
axes[1].tick_params(axis='x', rotation=45)


axes[2].plot(
    counts_top10.index.astype(str),
    counts_top10.values,
    marker='o',
    color='orange'
)
axes[2].set_title("Top 10 Merchant IDs (Line Chart)")
axes[2].set_xlabel("Merchant ID")
axes[2].set_ylabel("Number of Transactions")
axes[2].tick_params(axis='x', rotation=45)
axes[2].grid(True)

plt.tight_layout()
plt.savefig("merchant_dashboard.png", dpi=300, bbox_inches='tight')  # save
plt.show()

```
